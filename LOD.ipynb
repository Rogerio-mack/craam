```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
lod = pd.read_csv('https://github.com/Rogerio-mack/craam/raw/main/Delta_Omega.txt',sep='\\t',engine='python',skiprows=1,header=None)
lod.head()
```

```python
# check shift columns
lod.iloc[270:280].apply(lambda x: x.str.split())
```

```python
#check negative values
lod.iloc[14482:14490].apply(lambda x: x.str.split()) 
```

```python
lod = lod.apply(lambda x: x.str.split())
lod.head()
```

```python
year = []
month = []
day = []
dLod = []
sig = []

for i in range(len(lod)):
  year.append(lod.loc[i].values[0][0])
  month.append(lod.loc[i].values[0][1])
  day.append(lod.loc[i].values[0][2])
  dLod.append(lod.loc[i].values[0][3])
  sig.append(lod.loc[i].values[0][4])

LOD = pd.DataFrame()
LOD['year'] = year
LOD['month'] = month
LOD['day'] = day
LOD['dLod'] = dLod
LOD['sig'] = sig

LOD['year'] = LOD['year'].astype('int')
LOD['month'] = LOD['month'].astype('int')
LOD['day'] = LOD['day'].astype('int')
LOD['dLod'] = LOD['dLod'].astype('float32')
LOD['sig'] = LOD['sig'].astype('float32')

LOD.head()
```

```python
plt.figure(figsize=(20,5))

plt.plot(LOD.dLod)
plt.show()
```

```python
# check shift columns
lod.iloc[270:280].apply(lambda x: x.str.split())
```

```python
#check negative values
lod.iloc[14482:14490].apply(lambda x: x.str.split()) 
```

```python
lod = lod.apply(lambda x: x.str.split())
lod.head()
```

```python
year = []
month = []
day = []
dLod = []
sig = []

for i in range(len(lod)):
  year.append(lod.loc[i].values[0][0])
  month.append(lod.loc[i].values[0][1])
  day.append(lod.loc[i].values[0][2])
  dLod.append(lod.loc[i].values[0][3])
  sig.append(lod.loc[i].values[0][4])

LOD = pd.DataFrame()
LOD['year'] = year
LOD['month'] = month
LOD['day'] = day
LOD['dLod'] = dLod
LOD['sig'] = sig

LOD['year'] = LOD['year'].astype('int')
LOD['month'] = LOD['month'].astype('int')
LOD['day'] = LOD['day'].astype('int')
LOD['dLod'] = LOD['dLod'].astype('float32')
LOD['sig'] = LOD['sig'].astype('float32')

LOD.head()
```

```python
fig, ax1 = plt.subplots(figsize=(20,5))

ax1.plot(LOD.dLod,alpha=0.8,label='delta LOD',color='orange')
# ax1.plot(OMEGA.dLod,alpha=0.4)
plt.xticks(LOD[ (LOD['month'] == 1) & (LOD['day'] == 1)].index, labels=LOD[ (LOD['month'] == 1) & (LOD['day'] == 1)].year.astype('str'), rotation=60)
ax1.legend(loc='upper left')
ax2 = ax1.twinx()
ax2.plot(LOD.dLod.cumsum(),label='acumm LOD',alpha=0.8,color='k')

ax2.legend(loc='lower left')

plt.show()
```

```python
hydro = pd.read_csv('https://github.com/Rogerio-mack/craam/raw/main/hydro_power_generation_total.csv')
hydro = hydro[hydro.country.isin(['China','India'])]

for c in hydro.drop(columns='country'):
  if any(hydro[c].isnull()):
    hydro = hydro.drop(columns=c)

for c in hydro.drop(columns='country'):
  hydro[c] = hydro[c].replace({'k': '*1e3', 'M': '*1e6', 'B': '*1e9', 'TR': '*1e12'}, regex=True).map(pd.eval).astype(int)

hydro.head()
```

```python
gdp = pd.read_csv('https://github.com/Rogerio-mack/craam/raw/main/total_gdp_us_inflation_adjusted.csv')
gdp = gdp[gdp.country.isin(['China','India'])]

for c in gdp.drop(columns='country'):
  if any(gdp[c].isnull()):
    gdp = gdp.drop(columns=c)

for c in gdp.drop(columns='country'):
  gdp[c] = gdp[c].replace({'k': '*1e3', 'M': '*1e6', 'B': '*1e9', 'TR': '*1e12'}, regex=True).map(pd.eval).astype(int)

gdp.head()
```
